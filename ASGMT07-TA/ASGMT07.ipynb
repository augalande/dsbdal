```python
import re
doc=re.sub('[\W_]+',' ',doc)
doc
```

```python
word_token=nltk.word_tokenize(doc)
print(word_token)
```

```python
sent_token=nltk.sent_tokenize(doc)
print(sent_token)
```

```python
stopwords=set(nltk.corpus.stopwords.words('english'))
word_token=[token for token in word_token if token not in stopwords]
print(word_token)
```

```python
postag_token=nltk.pos_tag(word_token)
print(postag_token)
```

```python
lemmatizer=nltk.stem.WordNetLemmatizer()
lemmatize_token=[lemmatizer.lemmatize(token) for token in word_token]
print(lemmatize_token)
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_token=[stemmer.stem(token) for token in word_token]
print(stemmed_token)
```

```python
import numpy as np

def cal_tf(doc):
    word_token=nltk.word_tokenize(doc)
    num_token=len(word_token)
    tf={}
    unique_token,freqs=np.unique(word_token,return_counts=True)
    for token,freq in zip(unique_token,freqs):
        tf[token]=freq/num_token
    return tf

doc1=doc

doc_2=open('doc_02.txt','r')
doc2=doc_2.read()
doc_2.close()
doc2=re.sub('[\W_]+',' ',doc2)

tf1=cal_tf(doc1)
tf2=cal_tf(doc2)
print(tf1)
print()
print(tf2)
```

```python
import math

doc1_token=nltk.word_tokenize(doc1)
doc2_token=nltk.word_tokenize(doc2)

def cal_idf():
    N=2
    idf={}
    all_token=doc1_token+doc2_token
    for token in all_token:
        f=0
        if token in doc1_token:
            f+=1
        if token in doc2_token:
            f+=1
        idf[token]=math.log(N/f)
    return idf
idf=cal_idf()
print(idf)
```

```python
doc1_rep=[]
doc2_rep=[]

for token in doc1_token:
    doc1_rep.append(tf1[token]*idf[token])

for token in doc2_token:
    doc2_rep.append(tf2[token]*idf[token])

print(doc1_rep)
print()
print(doc2_rep)
        
```
