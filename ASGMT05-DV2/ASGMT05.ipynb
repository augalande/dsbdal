```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df=df.astype({"Name":"string","Sex":"string","Ticket":"string","Cabin":"string","Embarked":"string"})
```

```python
df.dtypes
```

```python
df["Age"]=df["Age"].fillna(df["Age"].median()) #use implace=True
```

```python
df["Fare"]=df["Fare"].fillna(df["Fare"].median())
```

```python
df["Cabin"]=df["Cabin"].fillna("Not Available")
```

```python
df.isnull().sum()
```

```python
df
```

```python
sns.boxplot(data=df,x="Sex",y="Age",hue="Survived")
```
